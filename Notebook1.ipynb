```python
# This code was reused from Code Reuse Session 1, Item 17
# function to find outliers
def find_outliers_IQR(data):

  q1=data.quantile(0.25)

  q3=data.quantile(0.75)

  IQR=q3-q1

  outliers = data[((data < (q1 - 1.5 * IQR)) | (data > (q3 + 1.5 * IQR)))]

  return outliers
```

```python
# This code was reused from Code Reuse Session 1, Item 19
# Finding outliers for Tumor Size
outliers = find_outliers_IQR(data['Tumor_Size'])

print("number of outliers: "+ str(len(outliers)))

outliers
```

```python
# This code was reused from Code Reuse Session 1, Item 19
# finding outliers for survival months
outliers = find_outliers_IQR(data['Survival_Months'])

print("number of outliers: "+ str(len(outliers)))

outliers
```

```python
# This code was reused from Code Reuse Session 1, Item 19
# finding outliers for age
outliers = find_outliers_IQR(data['Age'])

print("number of outliers: "+ str(len(outliers)))

outliers
```
